# 문서 요약

```python
# pdf 파일을 텍스트로 변환하기

import pymupdf
import os

pdf_file_path = "./data/인공지능 기법을 활용한 농촌지역의 객체 정보 추출방안.pdf"
doc = pymupdf.open(pdf_file_path)

text = ""
for page in doc:
    text += page.get_text()

print(text)
```

```python
# 요약을 위한 시스템 프롬프트를 생성한다.
system_prompt = f'''
너는 다음 글을 요약하는 봇이다. 아래 글을 읽고, 저자의 문제 인식과 주장을 파악하고, 주요 내용을 요약하라. 

작성해야 하는 포맷은 다음과 같다. 

# 제목

## 요약 (목적, 방법, 결과, 의의를 리스트로 정리한다.)

## 주요 키워드 (문서를 대표하는 키워드를 5개 이하로 추출한다.)

## 저자 소개 (이름, 소속, 직급, 연락처 정보를 리스트로 정리)


=============== 이하 텍스트 ===============

{ text }
'''

# OpenAI API를 사용하여 요약을 생성한다.
response = client.chat.completions.create(
    model='gpt-4o-mini',
    temperature=0.1,
    messages=[
        {"role": "system", "content": system_prompt},
    ]
)

summary = response.choices[0].message.content
print(summary)
```

```python
# 요약된 내용을 파일로 저장한다.

summary_file_path = "./data/output/인공지능_기법을_활용한_농촌지역의_객체_정보_추출방안_요약.txt"

os.makedirs(os.path.dirname(summary_file_path), exist_ok=True)

with open(summary_file_path, "w", encoding='utf-8') as f:
    f.write(summary)
```
